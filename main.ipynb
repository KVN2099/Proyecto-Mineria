```python
import pandas as pd
```

```python
df = pd.read_csv("nifty50.csv")
df.head()
```

```python
df_transformed = df.copy()
  
columns = ['open', 'high', 'low', 'close']
df_transformed[columns] = (df_transformed[columns] - df_transformed[columns].mean()) / df_transformed[columns].std()    

df_transformed.head()
```

```python
df_transformed.to_csv("nifty50-transformed.csv")
```

```python
df_new = df_transformed.copy()
df_new['year'] = df_new['date'].apply(lambda x: str(x)[:4])
df_new['month'] = df_new['date'].apply(lambda x: str(x)[4:6])
df_new['day'] = df_new['date'].apply(lambda x: str(x)[6:])
df_new.head()
```

```python
df_new.to_csv("nifty50-transformed.csv")
```
